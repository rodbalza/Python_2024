```python
with open('subvenciones.json', encoding='utf-8') as fich_lect, \
open('subvenciones_agrupadas.json', 'w', encoding='utf-8') as fich_escr:
    data = json.load(fich_lect)
    asoc_str = "Asociación"
    act_str = "Actividad Subvencionada"
    imp_str = "Importe en euros"
    lista = []
    lista_act = []
    asoc_actual = ""
    dicc = {}
    for elem in data:
        asoc = elem[asoc_str]
        act = elem[act_str]
        imp = elem[imp_str]
# Se extraen los valores correspondientes a las claves:
# asoc_str, act_str e imp_str del elemento actual y se 
# almacenan en las variables asoc, act e imp, respectivamente.
        if asoc_actual != asoc:            
            dicc["Actividades"] = lista_act
            dicc = {"Asociación": asoc}
            lista.append(dicc)
            lista_act = []
# Si asoc_actual no coincide con asoc, 
# se crea un nuevo diccionario dicc con la clave
# "Asociación" y se agrega a la lista: lista. 
# Luego, se reinicia la lista: lista_act
        lista_act.append({act_str : act, imp_str : imp})
# Se agrega un diccionario a lista_act con las claves:
# act_str y imp_str, y sus respectivos valores.
        asoc_actual = asoc
# Se actualiza asoc_actual con el valor de asoc.
    print(lista)
    json.dump(lista, fich_escr, ensure_ascii=False, indent=4) # , sort_keys=False
# finalmente, se escribe la lista: lista
# en el archivo “subvenciones_agrupadas.json” 
# con formato JSON, sin escapar caracteres especiales
# (ensure_ascii=False) y con sangría de 4 espacios (indent=4).
```
